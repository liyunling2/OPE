```python
import pandas as pd

path = "/Users/thaddeauslow/Desktop/IS455 Analytics in Asia/project/OPE/_5_GA_data/data_output/combined_restaurant_ga.parquet"

df = pd.read_parquet(path)

restaurant_list = (
    df[["restaurant_id", "name"]]
    .dropna()
    .drop_duplicates()
    .sort_values("name")
)

output_path = "/Users/thaddeauslow/Desktop/IS455 Analytics in Asia/project/OPE/_6_web_scraping/unique_restaurant_list.csv"

restaurant_list.to_csv(output_path, index=False)

print(f"Saved {len(restaurant_list)} unique restaurants to:")
print(output_path)
```
